```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('_data/HotelCustomersDataset.xlsx', engine='openpyxl')
```

```python
data = data.drop(columns=['NameHash', 'DocIDHash', 'ID'])
```

```python
data['Age'].mean(), data['Age'].median()
```

```python
sns.kdeplot(x='Age', data=data)
```

```python
data['Age'] = data['Age'].fillna(data['Age'].median())
```

```python
data.isna().sum()
```

```python
num_contires = 30
most_frequent_contries = list(data['Nationality'].value_counts().head(num_contires).index)

visitors = []
for i in range(len(most_frequent_contries)):
    visitors.append(len(data[data['Nationality'] == most_frequent_contries[i]]))

most_frequent_contries.append('others')
visitors.append(len(data) - np.sum(visitors))

plt.figure(figsize=(12, 12))
plt.pie(visitors, labels=most_frequent_contries, autopct='%1.1f%%');
```

```python
plt.figure(figsize=(20, 6))
plt.bar(x=most_frequent_contries, height=visitors)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
country_enc = LabelEncoder()
data['Nationality'] = country_enc.fit_transform(data['Nationality'])
```

```python
sns.countplot(x = 'DistributionChannel', data= data);
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x = 'MarketSegment', data= data);
```

```python
DC = pd.get_dummies(data['DistributionChannel'])

for col in DC.columns:
    data[col] = DC[col]
    
MS = pd.get_dummies(data['MarketSegment'])

for col in MS.columns:
    data[col] = MS[col]
```

```python
data = data.drop(columns=['DistributionChannel', 'MarketSegment'])
```

```python
data
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans

from sklearn.metrics import calinski_harabasz_score
```

```python
cls = GaussianMixture(n_components=4)
cls.fit(data)
```

```python
sns.countplot(cls.predict(data))
```

```python
from sklearn.decomposition import PCA
```

```python
X_2d = PCA(n_components=2).fit_transform(data)
```

```python
y = cls.predict(data)
```

```python
plt.scatter(X_2d[y==0, 0], X_2d[y==0, 1], alpha=0.1)
```

```python
plt.scatter(X_2d[y==1, 0], X_2d[y==1, 1], alpha=0.1)
```

```python
plt.scatter(X_2d[y==2, 0], X_2d[y==2, 1], alpha=0.1)
```

```python
plt.scatter(X_2d[y==3, 0], X_2d[y==3, 1], alpha=0.1)
```

```python
score = []
for n in range(2, 20):
    cls = GaussianMixture(n_components=n)
    cls.fit(data)
    score.append(calinski_harabasz_score(X=data, labels=cls.predict(data)))
    print(n, score[-1])
```
